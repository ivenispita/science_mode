## Linear fitting of random data  

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Random data of normal distribution, mean = 0, std = 0.2 and 0.5 and number of elements = 100
y = np.random.normal(loc=0.,scale=0.2, size=100)
x = np.random.normal(loc=0.,scale=0.5, size=100)
```

```python
# Regression coefficient 
Rxy = (np.std(x)/np.std(y))*np.corrcoef(x,y)[0,1]

# Model
xm = x.copy()
xm.sort()
model = xm*Rxy
```

## Plot example using xkdc (hand drawing style)

```python
with plt.xkcd():
    plt.plot(x,y,'.',markersize=12.,markerfacecolor='g',markeredgecolor='k',label='Data')
    plt.plot(xm,model,linestyle='solid',linewidth=5.5,color='r',label='Model')
    
    plt.legend()
    
    plt.title('Worst Model Ever ?',fontsize=25,fontweight='bold')
    plt.xlabel('X',fontsize=30,fontweight='bold')
    plt.ylabel('Y',fontsize=30,fontweight='bold')
    
    plt.tick_params(labelsize=20)
    
```
