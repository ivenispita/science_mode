## Loading SST from November 15$^{th}$ 1997 
### Source: Estimating the Circulation and Climate of the Ocean, Phase II (ECCO2): High-Resolution Global-Ocean and Sea-Ice Data Synthesis


```python

from netCDF4 import Dataset

# Loading data
url = 'http://ecco2.jpl.nasa.gov:80/opendap/data1/cube/cube92/lat_lon/quart_90S_90N/SST.nc/SST.1440x720.19971015.nc'
data = Dataset(url,mode='r')

```

```python

# Checking variables names 
data.variables.keys()
```

```python

# Acessing variables
latitude = data['LATITUDE_T'][:]
longitude = data['LONGITUDE_T'][:]-360.
sst = data['SST'][:].squeeze()
```

## Looking at the SST values

```python

import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline

# Quick look
plt.figure(figsize=(15,12))
plt.imshow(np.flipud(sst),aspect=True)
plt.colorbar()

plt.title("SST $^{\circ}$C",fontsize=30)

plt.show()
```
